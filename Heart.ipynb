```python
from sklearn.preprocessing import StandardScaler
# importing standard scaler for stanrdization
scaler = StandardScaler()
num_type= pd.DataFrame(scaler.fit_transform(num_types),
                                    columns = num_types.columns,
                                    index = num_types.index)
# concate after concate
num_types=pd.concat([num_type,temp],axis=1)
```

```python
X=num_types.drop('DEATH_EVENT',axis=1)
y=num_types['DEATH_EVENT']
# splitting dataset into 2 parts by separating target column
# impoorting libraries for model building purpose
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB

# importing evaluation metrics to know performances
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.model_selection import train_test_split
# splitting dataset into train and test dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
y_train.nunique()
# to know that target column is binary or odinal or multinomial
```

```python
alg_names=[]
alg_values=[]
#declaring the lists to store algorithm names and their accuracies
```

```python
# to print performances the author decalred a funtion below
def print_scores(alg, y_true, y_pred):
  print(alg)
  alg_names.append(alg)
  acc_score = accuracy_score(y_true, y_pred)
  print("accuracy: ",acc_score)
  alg_values.append(acc_score)
  pre_score = precision_score(y_true, y_pred)
  print("precision: ",pre_score)
  rec_score = recall_score(y_true, y_pred)
  print("recall: ",rec_score)
  f_score = f1_score(y_true, y_pred, average='weighted')
  print("f1_score: ",f_score)
```

```python
def modeling(alg, alg_name, params={}):
    model = alg(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print_scores(alg_name, y_test, y_pred)
    return model
# calling this funtion for fittiing and predicting purposes
```

```python
rf_model = modeling(RandomForestClassifier, "Random Forest Classification")
rf_model
# random forest algorithm
```

```python
dt_model = modeling(DecisionTreeClassifier, "Decision Tree Classification")
dt_model
# decision tree algorithm
```

```python
nb_model = modeling(GaussianNB, "Naive Bayes Classification")
nb_model
# naive bayes algorithm
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
gd_clf = GradientBoostingClassifier()
gd_clf.fit(X_train, y_train)

gd_pred = gd_clf.predict(X_test)
# prinitng the generated classification report
print_scores('Gradient Boosting', y_test, gd_pred)
# gradient boosting algorithm
```

```python
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(X_train, y_train)
# prinitng the generated classification report
gd_pred = knn_clf.predict(X_test)
print_scores('KNN', y_test, gd_pred)
# KNN algorithm
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
gd_clf = MLPClassifier()
gd_clf.fit(X_train, y_train)

gd_pred = gd_clf.predict(X_test)
# prinitng the generated classification reprot
print_scores('Neural Network', y_test, gd_pred)
# neural network
```
